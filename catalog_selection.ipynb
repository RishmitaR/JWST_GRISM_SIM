```python
from astropy.table import Table
from astropy.io import fits as pyfits
import astropy.units as u
from astropy.coordinates import SkyCoord,Angle

from regions import CircleSkyRegion, Regions

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import glob
```

```python
def load_cat( fi) : 
    d = Table.read(fi, format='fits')
    return(d)
```

```python
catalog = ("master_unicorn_v0.21_wallgrizli_v1.0_warcspec_wcigale_addew_addeazycolor.fits")
```

```python
df = load_cat(catalog).to_pandas()
```

```python
mass = 'bayes.stellar.m_star'

# Filter out invalid mass values before taking log
df_valid_mass = df[df[mass] > 0].copy()

# Convert valid mass values to log scale
df_valid_mass.loc[:, 'log_mass'] = np.log10(df_valid_mass[mass])

# Filter by redshift and log mass
df_sel = df_valid_mass.query("0.5 < REDSHIFT < 3.5 & log_mass > 9")

# Print columns to verify the selection
print(df_sel.columns)

# 0.5 < z < 1.00
df_sel_a = df_sel.query('0.5 < REDSHIFT < 1.00')

# 1.0 < z < 1.5
df_sel_b = df_sel.query('1.0 < REDSHIFT < 1.5')

# 1.5 < z < 2.0
df_sel_c = df_sel.query('1.5 < REDSHIFT < 2.0')

# 2.0 < z < 2.5
df_sel_d = df_sel.query('2.0 < REDSHIFT < 2.5')

# 2.5 < z < 3.5
df_sel_e = df_sel.query('2.5 < REDSHIFT < 3.5')
```

```python
#create UV and VJ columns
df_sel_a = df_sel_a.copy() 
df_sel_b = df_sel_b.copy() 
df_sel_c = df_sel_c.copy() 
df_sel_d = df_sel_d.copy() 
df_sel_e = df_sel_e.copy()

# create VJ columns in each DataFrame
df_sel_a.loc[:, 'VJ'] =  np.log10(df_sel_a['restV'])*-2.5 - np.log10(df_sel_a['restJ'])*-2.5 
df_sel_b.loc[:, 'VJ'] =  np.log10(df_sel_b['restV'])*-2.5 - np.log10(df_sel_b['restJ'])*-2.5 
df_sel_c.loc[:, 'VJ'] =  np.log10(df_sel_c['restV'])*-2.5 - np.log10(df_sel_c['restJ'])*-2.5 
df_sel_d.loc[:, 'VJ'] =  np.log10(df_sel_d['restV'])*-2.5 - np.log10(df_sel_d['restJ'])*-2.5 
df_sel_e.loc[:, 'VJ'] =  np.log10(df_sel_e['restV'])*-2.5 - np.log10(df_sel_e['restJ'])*-2.5 

# create UV columns in each DataFrame
df_sel_a.loc[:, 'UV'] =  np.log10(df_sel_a['restU'])*-2.5 - np.log10(df_sel_a['restV'])*-2.5 
df_sel_b.loc[:, 'UV'] =  np.log10(df_sel_b['restU'])*-2.5 - np.log10(df_sel_b['restV'])*-2.5 
df_sel_c.loc[:, 'UV'] =  np.log10(df_sel_c['restU'])*-2.5 - np.log10(df_sel_c['restV'])*-2.5 
df_sel_d.loc[:, 'UV'] =  np.log10(df_sel_d['restU'])*-2.5 - np.log10(df_sel_d['restV'])*-2.5 
df_sel_e.loc[:, 'UV'] =  np.log10(df_sel_e['restU'])*-2.5 - np.log10(df_sel_e['restV'])*-2.5 

# V-J arrays
V_J_a = np.asarray(df_sel_a.loc[:, 'VJ'])
V_J_b = np.asarray(df_sel_b.loc[:, 'VJ'])
V_J_c = np.asarray(df_sel_c.loc[:, 'VJ'])
V_J_d = np.asarray(df_sel_d.loc[:, 'VJ'])
V_J_e = np.asarray(df_sel_e.loc[:, 'VJ'])

# U-V arrays
U_V_a = np.asarray(df_sel_a.loc[:, 'UV'])
U_V_b = np.asarray(df_sel_b.loc[:, 'UV'])
U_V_c = np.asarray(df_sel_c.loc[:, 'UV'])
U_V_d = np.asarray(df_sel_d.loc[:, 'UV'])
U_V_e = np.asarray(df_sel_e.loc[:, 'UV'])
```

```python
# limit cut off

VJ = 'VJ'
UV = 'UV'

# 0 < z < 1.5
cutoff_a = df_sel_a.query(f"`{UV}` > 1.3 & `{VJ}` < 1.6")
cutoff_b = df_sel_b.query(f"`{UV}` > 1.3 & `{VJ}` < 1.6")

V_J_a_eq1 = np.sort(np.asarray(cutoff_a.loc[:, VJ]))
V_J_b_eq1 = np.sort(np.asarray(cutoff_b.loc[:, VJ]))

# 1.5 < z < 2.0
cutoff_c = df_sel_c.query(f"`{UV}` > 1.3 & `{VJ}` < 1.5")

V_J_c_eq2 = np.sort(np.asarray(cutoff_c.loc[:, VJ]))

# 2.0 < z < 3.0
cutoff_d = df_sel_d.query(f"`{UV}` > 1.2 & `{VJ}` < 1.4")
cutoff_e = df_sel_e.query(f"`{UV}` > 1.2 & `{VJ}` < 1.4")

V_J_d_eq3 = np.sort(np.asarray(cutoff_d.loc[:, VJ]))
V_J_e_eq3 = np.sort(np.asarray(cutoff_e.loc[:, VJ]))

# final equation array
U_V_a_eq = V_J_a_eq1 * 0.88 + 0.59
U_V_b_eq  = V_J_b_eq1 * 0.88 + 0.59
U_V_c_eq  = V_J_c_eq2 * 0.88 + 0.59
U_V_d_eq  = V_J_d_eq3 * 0.88 + 0.59
U_V_e_eq  = V_J_e_eq3 * 0.88 + 0.59
```

```python
#cutoff gen
def cutoff_gen(VJ,UV,lim):
    horizontal_x = np.linspace(0, VJ[0], num=50)
    horizontal_y = np.linspace(UV[0], UV[0], num=50)

    vertical_x = np.linspace(VJ[-1], VJ[-1], num=50)
    vertical_y = np.linspace(UV[-1], lim, num=50)

    return horizontal_x,horizontal_y,vertical_x,vertical_y
```

```python
#plotting
fig = plt.figure()
gs = fig.add_gridspec(2, 3, hspace=0.2, wspace=0)
axs = gs.subplots(sharex='col', sharey='row')

fig.set_figheight(8)
fig.set_figwidth(10)

horizontal_x_a,horizontal_y_a,vertical_x_a,vertical_y_a = cutoff_gen(V_J_a_eq1, U_V_a_eq,3)
horizontal_x_b,horizontal_y_b,vertical_x_b,vertical_y_b = cutoff_gen(V_J_b_eq1, U_V_b_eq,3)
horizontal_x_c,horizontal_y_c,vertical_x_c,vertical_y_c = cutoff_gen(V_J_c_eq2, U_V_c_eq,3)
horizontal_x_e,horizontal_y_e,vertical_x_e,vertical_y_e = cutoff_gen(V_J_e_eq3, U_V_e_eq,3)

axs[0, 0].scatter(V_J_a, U_V_a)
axs[0, 0].plot(V_J_a_eq1, U_V_a_eq, color = 'black')
axs[0, 0].plot(horizontal_x_a, horizontal_y_a, color = 'black')
axs[0, 0].plot(vertical_x_a, vertical_y_a, color = 'black')
axs[0, 0].set_title("0.5 < z < 1.0")


axs[0, 1].scatter(V_J_b, U_V_b)
axs[0, 1].plot(V_J_b_eq1, U_V_b_eq, color = 'black')
axs[0, 1].plot(horizontal_x_b, horizontal_y_b, color = 'black')
axs[0, 1].plot(vertical_x_b, vertical_y_b, color = 'black')
axs[0, 1].set_title("1.0 < z < 1.5")


axs[0, 2].scatter(V_J_c, U_V_c)
axs[0, 2].plot(V_J_c_eq2, U_V_c_eq, color = 'black')
axs[0, 2].plot(horizontal_x_c, horizontal_y_c, color = 'black')
axs[0, 2].plot(vertical_x_c, vertical_y_c, color = 'black')
axs[0, 2].set_title("1.5 < z < 2.0")


axs[1, 0].scatter(V_J_d, U_V_d)
axs[1, 0].plot(V_J_d_eq3, U_V_d_eq, color = 'black')
axs[1, 0].set_title("2.0 < z < 2.5")


axs[1, 1].scatter(V_J_e, U_V_e)
axs[1, 1].plot(V_J_e_eq3, U_V_e_eq, color = 'black')
axs[1, 1].plot(horizontal_x_e, horizontal_y_e, color = 'black')
axs[1, 1].plot(vertical_x_e, vertical_y_e, color = 'black')
axs[1, 1].set_title("2.5 < z < 3.5")

for ax in axs.flat:
    ax.set_ylim(0, 3)
    ax.set_yticks(np.arange(0, 2.5, 0.5))
    
for ax in axs[1, :]:
    ax.set_xlim(0, 2)
    ax.set_xticks(np.arange(0, 2.0, 0.5))

#axis titles
fig.text(0.5, 0.04, 'V-J', ha='center')
fig.text(0.04, 0.5, 'U-V', va='center', rotation='vertical')

fig.suptitle('NGDEEP Galaxies', fontsize=16)

plt.show()
```

```python
#quiesence selection

a_x_max = V_J_a_eq1[-1]
a_y_min = U_V_a_eq[0]

b_x_max = V_J_b_eq1[-1]
b_y_min = U_V_b_eq[0]

c_x_max = V_J_c_eq2[-1]
c_y_min = U_V_c_eq[0]

e_x_max = V_J_e_eq3[-1]
e_y_min = U_V_e_eq[0]

q_cat_a = df_sel_a.query(f"`{UV}` >= {a_y_min - 0.2} & `{VJ}` <= {a_x_max} & `{UV}` >= `{VJ}`*0.88 + 0.59")
q_cat_b = df_sel_b.query(f"`{UV}` >= {b_y_min} & `{VJ}` <= {b_x_max} & `{UV}` >= `{VJ}`*0.88 + 0.59")
q_cat_c = df_sel_c.query(f"`{UV}` >= {c_y_min - 0.2} & `{VJ}` <= {c_x_max} & `{UV}` >= `{VJ}`*0.88 + 0.59")
q_cat_e = df_sel_e.query(f"`{UV}` >= {e_y_min} & `{VJ}` <= {e_x_max} & `{UV}` >= `{VJ}`*0.88 + 0.59")

# V-J arrays quiescent
V_J_a_q = np.asarray(q_cat_a.loc[:, 'VJ'])
V_J_b_q = np.asarray(q_cat_b.loc[:, 'VJ'])
V_J_c_q = np.asarray(q_cat_c.loc[:, 'VJ'])
V_J_e_q = np.asarray(q_cat_e.loc[:, 'VJ'])

# U-V arrays quiescent
U_V_a_q = np.asarray(q_cat_a.loc[:, 'UV'])
U_V_b_q = np.asarray(q_cat_b.loc[:, 'UV'])
U_V_c_q = np.asarray(q_cat_c.loc[:, 'UV'])
U_V_e_q = np.asarray(q_cat_e.loc[:, 'UV'])
```

```python
#plotting quiesence

fig = plt.figure()
gs = fig.add_gridspec(2, 3, hspace=0.2, wspace=0)
axs = gs.subplots(sharex='col', sharey='row')

fig.set_figheight(8)
fig.set_figwidth(10)

horizontal_x_a,horizontal_y_a,vertical_x_a,vertical_y_a = cutoff_gen(V_J_a_eq1, U_V_a_eq,3)
horizontal_x_b,horizontal_y_b,vertical_x_b,vertical_y_b = cutoff_gen(V_J_b_eq1, U_V_b_eq,3)
horizontal_x_c,horizontal_y_c,vertical_x_c,vertical_y_c = cutoff_gen(V_J_c_eq2, U_V_c_eq,3)
horizontal_x_e,horizontal_y_e,vertical_x_e,vertical_y_e = cutoff_gen(V_J_e_eq3, U_V_e_eq,3)

axs[0, 0].scatter(V_J_a_q, U_V_a_q, color = 'red')
axs[0, 0].plot(V_J_a_eq1, U_V_a_eq, color = 'black')
axs[0, 0].plot(horizontal_x_a, horizontal_y_a, color = 'black')
axs[0, 0].plot(vertical_x_a, vertical_y_a, color = 'black')
axs[0, 0].set_title("0.5 < z < 1.0")


axs[0, 1].scatter(V_J_b_q, U_V_b_q, color = 'red')
axs[0, 1].plot(V_J_b_eq1, U_V_b_eq, color = 'black')
axs[0, 1].plot(horizontal_x_b, horizontal_y_b, color = 'black')
axs[0, 1].plot(vertical_x_b, vertical_y_b, color = 'black')
axs[0, 1].set_title("1.0 < z < 1.5")


axs[0, 2].scatter(V_J_c_q, U_V_c_q, color = 'red')
axs[0, 2].plot(V_J_c_eq2, U_V_c_eq, color = 'black')
axs[0, 2].plot(horizontal_x_c, horizontal_y_c, color = 'black')
axs[0, 2].plot(vertical_x_c, vertical_y_c, color = 'black')
axs[0, 2].set_title("1.5 < z < 2.0")


axs[1, 1].scatter(V_J_e_q, U_V_e_q, color = 'red')
axs[1, 1].plot(V_J_e_eq3, U_V_e_eq, color = 'black')
axs[1, 1].plot(horizontal_x_e, horizontal_y_e, color = 'black')
axs[1, 1].plot(vertical_x_e, vertical_y_e, color = 'black')
axs[1, 1].set_title("2.5 < z < 3.5")

for ax in axs.flat:
    ax.set_ylim(0, 3)
    ax.set_yticks(np.arange(0, 2.5, 0.5))
    
for ax in axs[1, :]:
    ax.set_xlim(0, 2)
    ax.set_xticks(np.arange(0, 2.0, 0.5))

#axis titles
fig.text(0.5, 0.04, 'V-J', ha='center')
fig.text(0.04, 0.5, 'U-V', va='center', rotation='vertical')

fig.suptitle('NGDEEP Quiescent Galaxies', fontsize=16)

plt.show()
```
